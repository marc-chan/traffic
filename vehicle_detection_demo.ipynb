## Building the tiny-YOLO model in Keras

In the previous demo on classification and localization, I used Tensorflow core to build, train and do predictions. Even for a small model (LeNet5), it still needed quite a bit of setup to be done e.g. Building the graph, Instantiating a Session, managing placeholders, variables etc...

Keras acts as a high level layer on top on Tensorflow, makes it simpler to get things setup and running. See how there is no need to set up constants and variables manually? Less code == less error :)

```python
####Model Parameters####
image_dims = (3,448,448)
p_dim = (2,2) #pooling size
f_dim = 3 #filter/kernel size
stride = (1,1) #stride size is (1,1) by default in keras conv layers
alpha = 0.1
########################

model = Sequential()
model.add(Conv2D(16,f_dim,input_shape=image_dims,padding="same"))
model.add(LeakyReLU(alpha=alpha))
model.add(MaxPooling2D(pool_size=p_dim))
model.add(Conv2D(32,f_dim,padding="same"))
model.add(LeakyReLU(alpha=alpha))
model.add(MaxPooling2D(pool_size=p_dim))
model.add(Conv2D(64,f_dim,padding="same"))
model.add(LeakyReLU(alpha=alpha))
model.add(MaxPooling2D(pool_size=p_dim))
model.add(Conv2D(128,f_dim,padding="same"))
model.add(LeakyReLU(alpha=alpha))
model.add(MaxPooling2D(pool_size=p_dim))
model.add(Conv2D(256,f_dim,padding="same"))
model.add(LeakyReLU(alpha=alpha))
model.add(MaxPooling2D(pool_size=p_dim))
model.add(Conv2D(512,f_dim,padding="same"))
model.add(LeakyReLU(alpha=alpha))
model.add(MaxPooling2D(pool_size=p_dim))
model.add(Conv2D(1024,f_dim,padding="same"))
model.add(LeakyReLU(alpha=alpha))
model.add(Conv2D(1024,f_dim,padding="same"))
model.add(LeakyReLU(alpha=alpha))
model.add(Conv2D(1024,f_dim,padding="same"))
model.add(LeakyReLU(alpha=alpha))
model.add(Flatten())
model.add(Dense(256))
model.add(Dense(4096))
model.add(LeakyReLU(alpha=alpha))
model.add(Dense(1470))
```

## Loading the pre-trained weights

In our previous session, we randomly initialized the layer weights, created a loss function to tell the model how to improve and fed the model training data across many epochs.

In practice however, training a convolutional network from scratch does not usually make sense due to insufficient training data / long time to train. Instead, we will load pre-trained weights (trained on huge datasets like ImageNet/COCO) and use that for our predictions.

### 1. Load from a numpy array

```python
### Loading the pretrained weights from a numpy array...
tiny_data = np.load("working_data/saved_models/yolo_tiny_weights.npy")

index = 0
for layer in model.layers:
    weights = layer.get_weights()
    if len(weights)>0:
        filter_shape, bias_shape = [w.shape for w in weights]
        if len(filter_shape)>2: #For convolutional layers
            filter_shape_i = filter_shape[::-1]
            bias_weight = tiny_data[index:index+np.prod(bias_shape)].reshape(bias_shape)
            index += np.prod(bias_shape)
            filter_weight= tiny_data[index:index+np.prod(filter_shape_i)].reshape(filter_shape_i)
            filter_weight= np.transpose(filter_weight,(2,3,1,0))
            index += np.prod(filter_shape)
            layer.set_weights([filter_weight,bias_weight])
            print("Layer {} intialized!".format(layer.name))
        else: #For hidden layers
            bias_weight = tiny_data[index:index+np.prod(bias_shape)].reshape(bias_shape)
            index += np.prod(bias_shape)
            filter_weight= tiny_data[index:index+np.prod(filter_shape)].reshape(filter_shape)
            index += np.prod(filter_shape)
            layer.set_weights([filter_weight,bias_weight])
            print("Layer {} intialized!".format(layer.name))
```

## Read in Data and Make Prediction

Notice how easy Keras makes inference of results. After the network is setup and weights initialized, all it takes is a single call to `model.predict(data)`

```python
## Define functions to zero-center data and convert zero-centered data back to RGB 255 values
def zero_center(data):
    data = ((data-(255/2))/(255/2)).astype(np.float32)
    return data

def zero_to_rgb(data):
    data = (data*(255/2)+(255/2)).astype(np.uint8)
    return data
```

```python
### Function to preprocess image and make prediction
def predict(model,image):
    data = zero_center(image)
    data = np.transpose(data,(2,0,1))
    if len(data.shape)<4:
        data = data[np.newaxis,:]
    result = model.predict(data)
    return result
```

```python
### Read and Feed image to tiny YOLO model
image = plt.imread("test_images/test1.jpg")
test_result = predict(model,image)
print("Shape of model output prediction tensor: {}".format(test_result.shape))
```

## Converting a Dense tensor into Sparse Predictions

The output of our model is a dense tensor with 1470 predicted values.

The YOLO model does detection by:
* Segmenting the input into a 7 x 7 grid, each cell responsible for predicting 1 object
* Predict 3 things for each grid cell:
    1. 20 Class probabilities (corresponds to classes in PASCAL VOC)
    2. 2 sets of Coordinates (x,y,width,height) and;
    3. A corresponding Confidence probability for each set of coordinates
* In total, 30 predictions for each grid cell (20 class + 2x4 coords + 2 confidence) 
* 30 x 7 x 7 = 1470

![tensor.png](img/tensor.png)


Obviously, not every cell contains an object (sparse predictions), therefore, we need to filter only for boxes that the model is most confident about.


```python
##### Extraction Parameters #####
grid_dims = 7
num_class = 20
num_coords = 2
num_confs = 2
target_class = {6:"Car"}
threshold = 0.05
#################################

### Function to extract bounding boxes
def extract_boxes(output,grid_dims=grid_dims,num_class=num_class,\
                  num_confs=num_confs,target_class=target_class,threshold=threshold):
    
    #######################################################################
    ## Step 1: Convert the model output into predictions for each grid cell
    #######################################################################
    num_boxes = grid_dims**2
    
    # Extract class probabilities for each grid cell
    class_probs = output[0:num_boxes*num_class].reshape([num_boxes,num_class])

    # Extract confidence probabilties that a grid cell contains an object
    confs = output[np.prod(class_probs.shape):np.prod(class_probs.shape)+num_boxes*num_confs].reshape([num_boxes,num_confs])

    # Extract predicted location coordinates of detected object
    coords = output[np.prod(class_probs.shape)+np.prod(confs.shape):].reshape([num_boxes,num_coords,4])
    
    #print("Shape of Class Probs: {}".format(class_probs.shape))
    #print("Shape of Confidences: {}".format(confs.shape))
    #print("Shape of coordinates: {}".format(coords.shape))
    
    #######################################################################
    ## Step 2: Filter predictions for detections with high probability
    #######################################################################
    boxes = []
    ## For each grid cell, extract one bounding box if threshold is met
    for cell in range(num_boxes):
        for b in range(num_confs):
            c =  confs[cell, b]
            x = (coords[cell, b, 0] + cell %  grid_dims) / grid_dims
            y = (coords[cell, b, 1] + cell // grid_dims) / grid_dims
            w =  coords[cell, b, 2] ** 2
            h =  coords[cell, b, 3] ** 2
            proba = class_probs[cell, :] * c
            
            p = 0
            cls = 0
            for cls_num in list(target_class.keys()):
                if proba[cls_num] >= threshold and proba[cls_num] > p:
                    p = proba[cls_num]
                    cls = cls_num
            if p > 0:
                boxes.append(np.array([x,y,w,h,p,cls]))
    return boxes
```

```python
boxes = extract_boxes(test_result[0])
plot_bounding(image,boxes)
```

## Filtering multiple predictions of the same object

When the position of an object cuts across grid lines, it is likely that multiple cells are predicting for the same object. To filter out these duplicate predictions, we will use the Non-Max Suppression technique. At its core, it's simply filtering out detected objects that have a high IOU with each other.

```python
def IOU(a,b):
    ## Step 1: unpack values
    tx, ty, tw, th, tp, tc = a
    px, py, pw, ph, pp, pc = b

    ## Step 2: If bounding boxes do not intersect, IOU = 0
    case_1 = ((tx-tw/2)>=(px+pw/2))or((ty-th/2)>=(py+ph/2))
    case_2 = ((tx+tw/2)<=(px-pw/2))or((ty+th/2)<=(py-ph/2))
    if case_1 or case_2:
        return 0

    ## Step 3: Calculate intersection
    bb_tlX = max(tx-tw/2,px-pw/2)
    bb_tlY = max(ty-th/2,py-ph/2)
    bb_brX = min(tx+tw/2,px+pw/2)
    bb_brY = min(ty+th/2,py+ph/2)
    intsec = (bb_brX-bb_tlX)*(bb_brY-bb_tlY)

    ## Step 4: Calculate union
    union = (tw*th)+(pw*ph)-intsec
        
    return intsec/union

def non_max_suppression(boxes,iou_threshold=0.05):
    
    ## Step 1: Sort predicted boxes according to their probability (desc.)
    sorted_boxes = sorted(boxes,key=lambda x:x[4],reverse=True)
    
    ## Step 2: Check IOU of boxes, set prob to 0 if IOU is high
    for i in range(len(sorted_boxes)):
        if sorted_boxes[i][4] == 0: 
            continue
        for j in range(i + 1, len(sorted_boxes)):
            if (IOU(sorted_boxes[i], sorted_boxes[j]) >= iou_threshold):
                sorted_boxes[j][4] = 0
                
    ## Step 3: Return the filtered boxes
    final_boxes = [box for box in sorted_boxes if box[4] > 0]
    return final_boxes
```

## Final results after performing Non-Max Supression

```python
final_boxes = non_max_suppression(boxes)
plot_bounding(image,final_boxes)
```

## Putting it all together

```python
## Bundle into a single function to do predictions
def predict_and_plot(model,filepath):
    img = Image.open(filepath)
    img = sq_center_crop(img)
    imgarr = np.asarray(img)
    result = predict(model,imgarr)[0]
    boxes = extract_boxes(result)
    final_boxes = non_max_suppression(boxes)
    plot_bounding(imgarr,final_boxes)
```
