```python
import requests
import pprint
import datetime
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import tensorflow as tf
import numpy as np
%matplotlib inline
from config import config


API_KEY = config["C_KEY"]
URL = "https://api.data.gov.sg/v1/transport/traffic-images"
```

```python
def convert_timestring(dt):
    #Format required for api query: YYYY-MM-DD[T]HH:MM:SS+08:00'
    return dt.strftime("%Y-%m-%dT%X+08:00")
```

```python
def get_one_traffic_data(date_time):
    r = requests.get(URL, params={"date_time": date_time}, headers={"api-key": API_KEY})
    data = r.json()
    return data
```

```python
def find_image_url(camera_id, data):
    camera_id = str(camera_id)
    c_arr = data["items"][0]["cameras"]
    for c in c_arr:
        if c["camera_id"] == camera_id:
            return c["image"]
    else:
        print("camera image not found")
```

```python
def pad_image(img):
    longer_side = max(img.size)
    horizontal_padding = (longer_side - img.size[0]) / 2
    vertical_padding = (longer_side - img.size[1]) / 2
    pad_img = img.crop(
        (
            -horizontal_padding,
            -vertical_padding,
            img.size[0] + horizontal_padding,
            img.size[1] + vertical_padding
        )
    )
    return pad_img
```

```python
#Define functions to zero-center data and convert zero-centered data back to RGB 255 values
def zero_center(data):
    data = (data.astype(np.float32)-(255/2))/(255/2)
    return data

def zero_to_rgb(data):
    data = (data*(255/2)+(255/2)).astype(np.uint8)
    return data
```

```python
def localize_image(camera_id,date_time):
    url = find_image_url(camera_id,get_one_traffic_data(date_time))
    response = requests.get(url, stream=True)
    img = pad_image(Image.open(BytesIO(response.content)))
    imgarr = np.reshape(zero_center(np.array(img)),(1,320,320,3))
    
    image_size = 320
    num_channels = 3
    num_reg_labels = 4
    num_labels = 2
    num_hidden = 128
    conv_size = 5
    conv_stride = 1
    pool_size = 2
    pool_stride = 2
    depth = 18
    batch_size = 1

    graph_lenet_demo = tf.Graph()

    with graph_lenet_demo.as_default():
        #set up variables/placeholders for tensorflow graph
        #demo_data = tf.placeholder(tf.float32, shape=(1,image_size,image_size,num_channels))
        demo_data = tf.constant(imgarr)

        #CONV_1
        conv1_weights = tf.constant(c1w)
        conv1_bias = tf.constant(c1b)

        #CONV_2
        conv2_weights = tf.constant(c2w)
        conv2_bias = tf.constant(c2b)

        #FC_1: layer mapping to 128 hidden nodes (for classification)
        fc1_weights = tf.constant(fc1w)
        fc1_bias = tf.constant(fc1b)

        #FC_2: layer mapping to 2 logits (for classification)
        fc2_weights = tf.constant(fc2w)
        fc2_bias = tf.constant(fc2b)

        #FC_1b: layer mapping to 128 hidden nodes (for localization)
        fc1b_weights = tf.constant(fc1bw)
        fc1b_bias = tf.constant(fc1bb)

        #FC_2b: layer mapping to 4 coordinates(for localization)
        fc2b_weights = tf.constant(fc2bw)
        fc2b_bias = tf.constant(fc2bb)

        def model(data):
            conv1 = tf.nn.conv2d(data,filter=conv1_weights,strides=[1,conv_stride,conv_stride,1],padding="SAME")+conv1_bias
            conv1_relu = tf.nn.relu(conv1)
            conv1_pool = tf.nn.max_pool(conv1_relu,[1,pool_size,pool_size,1],[1,pool_stride,pool_stride,1],padding="SAME")

            conv2 = tf.nn.conv2d(conv1_pool,filter=conv2_weights,strides=[1,conv_stride,conv_stride,1],padding="SAME")+conv2_bias
            conv2_relu = tf.nn.relu(conv2)
            conv2_pool = tf.nn.max_pool(conv2_relu,[1,pool_size,pool_size,1],[1,pool_stride,pool_stride,1],padding="SAME")

            conv2_flat = tf.reshape(conv2_pool,[conv2_pool.get_shape().as_list()[0],-1])

            fc_l1 = tf.nn.relu(tf.matmul(conv2_flat,fc1_weights)+fc1_bias)
            fc_l1b = tf.nn.relu(tf.matmul(conv2_flat,fc1b_weights)+fc1b_bias)

            logits = tf.matmul(fc_l1,fc2_weights)+fc2_bias
            label = tf.argmax(tf.nn.softmax(logits),axis=1)
            coords = tf.matmul(fc_l1b,fc2b_weights)+fc2b_bias
            
            return label, coords

        tlabel, tcoords = model(demo_data)
        
    with tf.Session(graph=graph_lenet_demo) as sess:
        #fd = {"demo_data":imgarr}
        label, coords = sess.run([tlabel, tcoords])

    
    fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,sharex=True,sharey=True,figsize=(10,5))
    fig.suptitle("Traffic Camera "+str(camera_id)+" Image at: "+date_time, fontsize=16)
    ax1.set_title("Original")
    ax1.imshow(img)
    ax2.set_title("Prediction: "+str(label[0]))
    ax2.imshow(img)
    if str(label[0]) != "0":
        tlX = coords[0][0]
        tlY = coords[0][1]
        brX = coords[0][2]
        brY = coords[0][3]
        rect = patches.Rectangle((tlX,tlY),brX-tlX,brY-tlY,linewidth=1,edgecolor='r',facecolor='none')
        ax2.add_patch(rect)
```

```python
image_size = 320
num_channels = 3
num_reg_labels = 4
num_labels = 2
num_hidden = 128
conv_size = 5
conv_stride = 1
pool_size = 2
pool_stride = 2
depth = 18
batch_size = 20
class_loss_split = 0.5

graph_lenet_local = tf.Graph()

with graph_lenet_local.as_default():
    #set up variables/placeholders for tensorflow graph
    tf_train_data = tf.placeholder(tf.float32, shape=(batch_size,image_size,image_size,num_channels))
    tf_train_labels = tf.placeholder(tf.float32, shape=(batch_size,num_labels))
    tf_train_reg_labels = tf.placeholder(tf.float32, shape=(batch_size,num_reg_labels))
    #tf_val_data = tf.constant(val_data)
    #tf_val_labels = tf.constant(hot_val_labels)
    #tf_val_reg_labels = tf.constant(val_reg_labels)
    
    #CONV_1
    conv1_weights = tf.Variable(tf.truncated_normal([conv_size,conv_size,num_channels,depth],stddev=0.1))
    conv1_bias = tf.Variable(tf.zeros([depth]))
    
    #CONV_2
    conv2_weights = tf.Variable(tf.truncated_normal([conv_size,conv_size,depth,depth],stddev=0.1))
    conv2_bias = tf.Variable(tf.zeros([depth]))
    
    #FC_1: layer mapping to 128 hidden nodes (for classification)
    fc1_weights = tf.Variable(tf.truncated_normal([image_size//pool_size//pool_size*image_size//pool_size//pool_size*depth,num_hidden],stddev=0.1))
    fc1_bias = tf.Variable(tf.zeros([num_hidden]))
    
    #FC_2: layer mapping to 2 logits (for classification)
    fc2_weights = tf.Variable(tf.truncated_normal([num_hidden,num_labels],stddev=0.1))
    fc2_bias = tf.Variable(tf.zeros([num_labels]))
    
    #FC_2b: layer mapping to 4 coordinates(for localization)
    fc2b_weights = tf.Variable(tf.truncated_normal([image_size//pool_size//pool_size*image_size//pool_size//pool_size*depth,num_reg_labels],stddev=0.1))
    fc2b_bias = tf.Variable(tf.zeros([num_reg_labels]))
    
    def model(data):
        conv1 = tf.nn.conv2d(data,filter=conv1_weights,strides=[1,conv_stride,conv_stride,1],padding="SAME")+conv1_bias
        conv1_relu = tf.nn.relu(conv1)
        conv1_pool = tf.nn.max_pool(conv1_relu,[1,pool_size,pool_size,1],[1,pool_stride,pool_stride,1],padding="SAME")
        
        conv2 = tf.nn.conv2d(conv1_pool,filter=conv2_weights,strides=[1,conv_stride,conv_stride,1],padding="SAME")+conv2_bias
        conv2_relu = tf.nn.relu(conv2)
        conv2_pool = tf.nn.max_pool(conv2_relu,[1,pool_size,pool_size,1],[1,pool_stride,pool_stride,1],padding="SAME")
        
        conv2_flat = tf.reshape(conv2_pool,[conv2_pool.get_shape().as_list()[0],-1])

        fc_l1 = tf.nn.relu(tf.matmul(conv2_flat,fc1_weights)+fc1_bias)
        
        logits = tf.matmul(fc_l1,fc2_weights)+fc2_bias
        coords = tf.matmul(conv2_flat,fc2b_weights)+fc2b_bias
        return logits, coords
    
    train_logits, train_coords = model(tf_train_data)
    train_class_loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=tf_train_labels,logits=train_logits))
    train_reg_loss = tf.reduce_mean(tf.sqrt(tf.square(train_coords-tf_train_reg_labels)))
    train_loss = (class_loss_split*train_class_loss)+((1-class_loss_split)*train_reg_loss)
    
    optimizer = tf.train.AdamOptimizer(0.00001).minimize(train_loss)
    
    #val_logits, val_coords = model(tf_val_data)
    #val_class_loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=tf_val_labels,logits=val_logits))
    #val_reg_loss = tf.reduce_mean(tf.sqrt(tf.square(val_coords-tf_val_reg_labels)))
    #val_loss = (class_loss_split*val_class_loss)+((1-class_loss_split)*val_reg_loss)
    
    #val_pred_label = tf.nn.softmax(val_logits)
    #val_pred_reg = val_coords
    
    saver = tf.train.Saver()
    
#Extract pretrained classification labels
with tf.Session(graph=graph_lenet_local) as sess:
    lenet_loss_history = []
    saver.restore(sess, "working_data/saved_models/lenet_local.ckpt")
    #CONV_1
    c1w = conv1_weights.eval()
    c1b = conv1_bias.eval()
    
    #CONV_2
    c2w = conv2_weights.eval()
    c2b = conv2_bias.eval()
    
    #FC_1: layer mapping to 128 hidden nodes (for classification)
    fc1w = fc1_weights.eval()
    fc1b = fc1_bias.eval()
    
    #FC_2: layer mapping to 2 logits (for classification)
    fc2w = fc2_weights.eval()
    fc2b = fc2_bias.eval()
    
del graph_lenet_local
```
