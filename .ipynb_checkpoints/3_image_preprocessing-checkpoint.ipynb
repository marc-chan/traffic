### 3. Image Preprocessing

This notebook does preprocessing on the image files downloaded in the previous notebook. Padding, mirroring and splitting into train/validation sets (80:20).

```python
from PIL import Image, ImageOps
import os
import numpy as np
import pickle
```

```python
#####Parameters#####
img_size = 320
channels = 3
samples = 500
###################
```

```python
def pad_image(img):
    longer_side = max(img.size)
    horizontal_padding = (longer_side - img.size[0]) / 2
    vertical_padding = (longer_side - img.size[1]) / 2
    pad_img = img.crop(
        (
            -horizontal_padding,
            -vertical_padding,
            img.size[0] + horizontal_padding,
            img.size[1] + vertical_padding
        )
    )
    return pad_img

def pad_folder(camera_id):
    camera_id = str(camera_id)
    input_path = os.getcwd()+"/images/"+camera_id
    output_path = os.getcwd()+"/images/padded/"+camera_id
    os.makedirs(output_path, exist_ok=True)
    input_file_names = os.listdir(input_path)
    counter = 0
    for file in input_file_names:
        img = Image.open(input_path+"/"+file)
        pad_img = pad_image(img)
        pad_img.save(output_path+"/"+file)
        counter += 1
    print("Padding of {} images done".format(counter))
```

```python
#display image before and after padding
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
img_before = Image.open("images/raw/1502/1502-45.jpg")
img_after = Image.open("images/padded/1502/1502-45.jpg")
fig, ax = plt.subplots(nrows=1, ncols=2,sharex=True,)
ax[0].imshow(img_before)
ax[0].set_title("Before Padding")
ax[1].imshow(img_after)
ax[1].set_title("After Padding")
```

```python
def compile_image_data(camera_id,label,img_size,tagrange,number,):
    print("Starting compilation...")
    camera_id = str(camera_id)
    input_path = os.getcwd()+"/images/localization/"+camera_id+"/"+str(label)
    input_file_names = [camera_id+"-"+str(x)+".jpg" for x in range(tagrange)]
    data = np.ndarray((number,img_size,img_size,channels),dtype=np.uint8)
    counter = 0
    for i,file in enumerate(input_file_names):
        try:
            img = Image.open(input_path+"/"+file)
        except:
            continue
        img_arr = np.array(img, dtype='uint8')
        data[counter,:,:,:] = img_arr
        counter += 1
    print("Compilation done with {} items!".format(counter))
    return data
```

```python
def create_mirror(images):
    data = np.ndarray((images.shape[0],images.shape[1],images.shape[2],images.shape[3]),dtype=np.uint8)
    for i,image in enumerate(images):
        mirror = np.array(ImageOps.mirror(Image.fromarray(image)), dtype='uint8')
        data[i,:,:,:] = mirror
    return data
```

```python
positive = compile_image_data(1502,1,img_size,8000,samples)
negative = np.random.permutation(compile_image_data(1502,0,img_size,8000,1830))[0:samples]
positive_flip = create_mirror(positive)
negative_flip = create_mirror(negative)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,sharex=True)
ax[0].imshow(positive[10])
ax[0].set_title("Original")
ax[1].imshow(positive_flip[10])
ax[1].set_title("Mirror Image")
```

```python
import csv
with open("images/localization/1502/localization_labels.csv","r") as f:
    csv_file = csv.reader(f)
    positive_labels = [l for l in csv_file]
    
positive_labels=np.array(positive_labels[1:],dtype=np.float32)
positive_labels[:,0] = 1

positive_flip_labels = positive_labels.copy()
tlX_flip = img_size - positive_flip_labels[:,3]
brX_flip = img_size - positive_flip_labels[:,1]
positive_flip_labels[:,1] = tlX_flip
positive_flip_labels[:,3] = brX_flip
```

```python
negative_labels = np.zeros((negative.shape[0],5))
negative_flip_labels = np.zeros((negative_flip.shape[0],5))
```

```python
data = np.vstack((positive,positive_flip,negative,negative_flip))
labels = np.vstack((positive_labels,positive_flip_labels,negative_labels,negative_flip_labels))
```

```python
def convert_bb(im_w, im_h, tlX, tlY, brX, brY):
    x = (brX+tlX)/2/im_w
    y = (brY+tlY)/2/im_h
    w = np.sqrt((brX-tlX)/im_w)
    h = np.sqrt((brY-tlY)/im_h)
    return x, y, w, h
```

```python
for i,l in enumerate(labels):
    labels[i,1:] = convert_bb(img_size,img_size,l[1],l[2],l[3],l[4])
```

```python
dump = {"data": data, "labels": labels}
pickle.dump(dump, open(os.getcwd()+"/setup_data/1502_data.p","wb"),pickle.HIGHEST_PROTOCOL)
```

```python
imgs = pickle.load(open(os.getcwd()+"/setup_data/1502_data.p","rb"))

plt.imshow(imgs["data"][10])
```

```python
#Split into training/validation (80:20) sets
data = imgs["data"]
labels = imgs["labels"]

np.random.seed(42)
val_index = np.random.choice(data.shape[0],400,replace=False)
train_index_mask = np.invert(np.in1d(np.arange(data.shape[0]), val_index))
train_data = data[train_index_mask]
train_labels = labels[train_index_mask]
val_data = data[val_index]
val_labels = labels[val_index]
data_dump = {
    "train_data": train_data,
    "train_labels": train_labels,
    "val_data": val_data,
    "val_labels": val_labels
}
pickle.dump(data_dump,open("working_data/working_set.p","wb"))

del imgs
del data
del labels
del data_dump
```
